### Pregunta 1 ###

Tenemos que la funcion error definida de la siguiente forma:
            $$erf(x)=\frac{2}{\pi}\int_{0}^{x}e^{-t^{2}}dt$$

El metodo trapezoidal se define de la siguiente manera:
                $$I=\frac{h}{2}(f(x_0)+2f(x_1)+...+2f(x_{n-1})+f(x_n))$$

```python
import scipy.special
plt.style.use('seaborn-poster')
```

```python
def errf(x):
    valor=np.sqrt(2/(np.pi))
    t=pow(x,2)
    f=np.power(np.e,-t)
    return f*valor
```

Definimos la funcion que ejecuta la regla trapezoidal

```python
def trapezoidal(a,b,n):
    c1=a
    h=(b-a)/n
    suma=0.0
    y=np.zeros(n)
    x_0=errf(a)
    y[0]=x_0
    for i in range(n-2):
        c1=c1+h
        valor=errf(c1)
        suma=suma+valor*2
        y[i]=suma
    x_n=errf(b)
    total=x_0+x_n+suma
    total=total*(h/2)
    y[n-1]=total
    return y
```
